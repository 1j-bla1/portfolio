```python
%%javascript

class Dice { // What should the name of the class be?
  constructor(sides) { // Constructor to initialize the number of sides
    this.sides = sides; // Number of sides on the dice
  }

  roll() {
    return Math.floor(Math.random() * this.sides) + 1; // Rolls the dice and returns a number between 1 and the number of sides
  }
}
let sides = 6; // Number of sides for the dice
// Test the Dice class:
// INSTANTIATE A NEW DICE OBJECT WITH A NUMBER OF SIDES
const dice = new Dice(sides); // Create a 6-sided dice
// CALL THE ROLL METHOD A FEW TIMES AND PRINT THE RESULTS
console.log("Number of sides: " + dice.sides);
console.log("Roll 1: " + dice.roll());
console.log("Roll 2: " + dice.roll());
console.log("Roll 3: " + dice.roll());
// CALL THE ROLL METHOD A FEW TIMES AND PRINT THE RESULTS
```

```python
%%javascript

// Utility function to log to the output area of this cell (not the whole document)
function logOutput(message) {
    var outputDiv = document.getElementById("cell-output");
    if (!outputDiv) {
        outputDiv = document.createElement("div");
        outputDiv.id = "cell-output";
        outputDiv.style.fontFamily = "monospace";
        outputDiv.style.whiteSpace = "pre";
        outputDiv.style.marginTop = "10px";
        this.element.append(outputDiv);
    }
    var p = document.createElement("div");
    p.textContent = message;
    outputDiv.appendChild(p);
}

//1.Person Class
class Person {
    constructor(name, age) {
        this.name = name;
        this.age = age;
    }
    //2.Greet
    greet() {
        logOutput("Hello! My name is " + this.name);
    }
}

//3. Book Class
class Book {
    constructor(title, author, pages) {
        this.title = title;
        this.author = author;
        this.pages = Number(pages); // ensure it's a number
        this.pagesRead = 0;         // initialize pagesRead
    }
    //4. Pages
    read(pages) {
        let pagesToRead = Number(pages);
        this.pagesRead += pagesToRead;
        if (this.pagesRead > this.pages) {
            this.pagesRead = this.pages; // don't exceed total pages
        }
        logOutput(`${pagesToRead} pages read. Total pages read: ${this.pagesRead}/${this.pages}`);
    }
    // Step 5: LongBook method
    isLongBook() {
        return this.pages > 300;
    }
}

// Step 6: Car class
class Car {
    constructor(make, model, fuel) {
        this.make = make;
        this.model = model;
        this.fuel = Number(fuel);
    }
    // Drive method
    drive(distance) {
        let fuelConsumed = distance * 0.1;
        if (fuelConsumed > this.fuel) {
            logOutput(`${this.make} ${this.model} cannot drive ${distance} km. Not enough fuel.`);
        } else {
            this.fuel -= fuelConsumed;
            logOutput(`${this.make} ${this.model} drove ${distance} km. Remaining fuel: ${this.fuel.toFixed(2)}`);
        }
    }
    // Refuel method
    refuel(amount) {
        this.fuel += Number(amount);
        logOutput(`${this.make} ${this.model} refueled by ${amount}. Total fuel: ${this.fuel.toFixed(2)}`);
    }
    // Step 7: static method
    static info() {
        logOutput("Cars are vehicles.");
    }
}

// Step 8: ElectricCar class
class ElectricCar extends Car {
    constructor(make, model, fuel, battery) {
        super(make, model, fuel);
        this.battery = Number(battery);
    }
    charge(amount) {
        this.battery += Number(amount);
        if (this.battery > 100) this.battery = 100;
        logOutput(`${this.make} ${this.model} charged by ${amount}%. Battery level: ${this.battery}%`);
    }
}

//Test Person Class
let person1 = new Person("James", 14);

//Test Book Class
let myBook = new Book("The Hobbit", "J.R.R. Tolkien", 310);

//Printing Person Object 
logOutput("Person object: " + JSON.stringify(person1));
logOutput("Name: " + person1.name);
logOutput("Age: " + person1.age);

//Printing Book Object
logOutput("Book object: " + JSON.stringify(myBook));
logOutput("Title: " + myBook.title);
logOutput("Author: " + myBook.author);
logOutput("Pages: " + myBook.pages);

//Printing Greet
person1.greet();

// Read some pages
myBook.read(50);
myBook.read(100);

//Return true if over 300 pages
logOutput("Is this a long book? " + myBook.isLongBook());

// Test the Car class
let myCar = new Car("Toyota", "Corolla", 50);
myCar.drive(100);
myCar.refuel(20);

// Static method print
Car.info();

// Output for electriccar
let myElectricCar = new ElectricCar("Tesla", "Model S", 0, 50);
myElectricCar.charge(30);
myElectricCar.charge(50);

// --- Step 10: Numbered Combined Routine ---
function runRoutine() {
    logOutput("===== Step 10: Numbered Combined Routine =====");

    // Create objects
    let person1 = new Person("James", 14);
    let myBook = new Book("The Hobbit", "J.R.R. Tolkien", 310);
    let myCar = new Car("Toyota", "Corolla", 50);
    let myElectricCar = new ElectricCar("Tesla", "Model S", 0, 50);

    // Step 1: Person Action
    logOutput("1) Person Action:");
    logOutput("   Name: " + person1.name);
    logOutput("   Age: " + person1.age);
    person1.greet();

    // Step 2: Book Action
    logOutput("2) Book Action:");
    logOutput("   Title: " + myBook.title);
    logOutput("   Author: " + myBook.author);
    logOutput("   Pages: " + myBook.pages);
    myBook.read(30);
    logOutput("   Is this a long book? " + myBook.isLongBook());

    // Step 3: Car Action
    logOutput("3) Car Action:");
    myCar.drive(50);
    myCar.refuel(10);

    // Step 4: Electric Car Action
    logOutput("4) Electric Car Action:");
    myElectricCar.charge(20);

    // Step 5: Static Car Info
    logOutput("5) Car Class Info:");
    Car.info();

    logOutput("===== End of Routine =====");
}

// --- Run the numbered routine ---
runRoutine();
```
